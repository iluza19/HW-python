```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv('kc_house_data.csv', sep = ',', encoding='latin-1')
df.head()
```

```python
plt.hist(df['price'], color = 'blue', edgecolor = 'black',
         bins = int(180/5))
plt.title('Стоимость квартир')
plt.xlabel('стоимость')
plt.ylabel('количество')
```

```python
data = df['sqft_living'].value_counts()
plt.bar(data.index, data.values)
plt.xticks(rotation=50)
```

```python
data = df['yr_built'].value_counts()
plt.bar(data.index, data.values)
plt.xticks(rotation=50)
```

Выводы: Большинство квартир стоят ниже миллиона, большинство домов было построено после 2010 года.
